```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
df = pd.read_csv('data.csv')
df = df*-1
```

```python
#NEWS
rows = df.iloc[:,1].size
MIN = -75.0
MAX = -45.0
for i in range(6):
    res = df.iloc[:,i]
    c = res.copy()
    c[c<MIN] = MIN
    c[c>MAX] = MAX
    c = (c-MIN)/(MAX-MIN)
    plt.scatter(x=df.index, y=np.ones(rows)*i,s=abs(res),c=c,alpha=0.6)
ca = plt.gca()

rect = []
rect.append(patches.Rectangle((0,4.2),100,0.7,linewidth=1,edgecolor='r',facecolor='none'))
rect.append(patches.Rectangle((0,1.2),50,2.7,linewidth=1,edgecolor='r',facecolor='none'))
rect.append(patches.Rectangle((53,1.2),47,2.7,linewidth=1,edgecolor='r',facecolor='none'))
rect.append(patches.Rectangle((0,0.2),100,0.7,linewidth=1,edgecolor='r',facecolor='none'))

for rt in rect:
    ca.add_patch(rt)
ca.text(50,4.5,'Building One',horizontalalignment='center')
ca.text(50,0.5,'Building Two',horizontalalignment='center')
ca.text(75,2.5,'Building Main',horizontalalignment='center')
ca.text(25,2.5,'Square',horizontalalignment='center')
plt.colorbar()

ca.arrow(105,4,0,1,length_includes_head=True,head_width=3, head_length=0.5, fc='k', ec='k')
ca.text(105,3.6,'North',horizontalalignment='center', rotation='vertical')

plt.xlim(0,110)
plt.title('Geographical distribution of field strength')

```
